```python
flat = fits.getdata("/home/carterrhea/Documents/OMM-Clean/241109/master_flat.fits").astype(np.float32)
#data = fits.getdata("/home/carterrhea/Documents/OMM-Data/241109/Target/M74_pos1/241109_0000000915.fits").astype(np.float32)

files = glob("/home/carterrhea/Documents/OMM-Data/241109/Target/M74_pos1/*.fits")

```
